```python
import midpoint
import time
import csv
```

```python
# Labeling model (grids)

# DEFINE INPUT

# Dimension of square grid
dim = 20

# Number of districts
num = 2

# Max allowed pop deviation (1% => 0.01)
balance_epsilon = 0.02

# Beta
beta = 0.5

# File names for plans A & B and grid shapefile
filename_A = 'Grid_Plans/PlanA_'+str(dim)+'x'+str(dim)+'_'+str(num)+'.csv'
filename_B = 'Grid_Plans/PlanB_'+str(dim)+'x'+str(dim)+'_'+str(num)+'.csv'
shapefile = 'GridShapefiles/'+str(dim)+'x'+str(dim)+'/'+str(dim)+'x'+str(dim)+'.shp'

# Desired file name for midpoint plan (plus extension) and midpoint figure (no extension)
filename_mid = 'Betapoint_A_B_'+str(dim)+'x'+str(dim)+'_'+str(num)+'_Beta'+str(beta)+'.csv'
figname_mid = 'Betapoint_A_B_fig_'+str(dim)+'x'+str(dim)+'_'+str(num)+'_Beta'+str(beta)

# Indicate whether or not to obtain a heuristic warm-start (via endpoint plan or local search)
warm_start_Init = True
warm_start_LS = False

# Indicate whether or not to optimize for compactness after midpoint
compact_bool = True

# Column to weight by
col_weight = 'UNWEIGHTED'

# Indicate whether grid shapefile or not
grid_bool = True



# RUN PROGRAM

# Record start time
start_cpu = time.process_time()
start_wall = time.time()

# Determine (possibly approximate) betapoint
TD,opt_val,midpoint_plan_assignment,new_A,actual_beta = midpoint.TD_midpoint(shapefile,filename_A,filename_B,balance_epsilon,
                                                                             warm_start_LS,warm_start_Init,compact_bool,
                                                                             col_weight,beta,grid_bool)

# Report run time
print('Runtime (cpu): ',time.process_time()-start_cpu,' seconds')
print('Runtime (wall): ',time.time()-start_wall,' seconds')

# Output and visualize the result
midpoint.output_and_visualize_result(midpoint_plan_assignment,new_A,filename_B,shapefile,filename_mid,figname_mid)
```

```python
# Labeling model (grids -- looped for paper examples)

# DEFINE INPUT

# Dimension of square grid
dim = 24

# Number of districts
num = 4

# Column string to weight by ('UNWEIGHTED' for uniform/unweighted case)
col_weight = 'UNWEIGHTED'

# Indicate whether grid shapefile or not
grid_bool = True

# Max allowed deviation from average part weight (1% => 0.01)
if dim == 24 and num == 4:
    balance_epsilon = 0.02
elif dim == 24 and num == 8:
    balance_epsilon = 0.03
elif dim == 40 and num == 4:
    balance_epsilon = 0.006
else:
    balance_epsilon = 0.012
    
# Beta-values and types of experiments
#beta_list = [0.5,0.25,0.75]
beta_list = [0.5]
ex_list = ['Nothing']
#ex_list = ['Nothing','Warm','Nothing-Comp','Comp']


# Run
for beta in beta_list:
    for ex in ex_list:
        
        if ex == 'Nothing':
            
            compact_bool = False
            warm_start_Init = False
            warm_start_LS = False
            
        elif ex == 'Warm':
            
            compact_bool = False
            warm_start_Init = True
            warm_start_LS = False
            
        elif ex == 'Nothing-Comp':
            
            compact_bool = True
            warm_start_Init = False
            warm_start_LS = False
            
        else:
            
            compact_bool = True
            warm_start_Init = True
            warm_start_LS = False



        # Desired file name for run times and distance values
        outfile = open('Results_'+str(dim)+'x'+str(dim)+'_'+str(num)+'_Beta'+str(beta)+'_'+ex+'.csv','w')

        writer = csv.writer(outfile,delimiter=',')
        writer.writerow(['Plan A','Plan B','t1(A,B)','CPU time','Wall time'])

        # Iterate
        for k in range(1,21):

            temp = 25*k

            print('\n')
            print(temp,' and ',temp+500)

            # Filenames for plan A, plan B, and shapefile
            filename_A = 'RandomPartitions_Grids/'+str(dim)+'x'+str(dim)+'_'+str(num)+'/Random_'+str(dim)+'x'+str(dim)+'dim_'+str(num)+'part_'+str(temp)+'.csv'
            filename_B = 'RandomPartitions_Grids/'+str(dim)+'x'+str(dim)+'_'+str(num)+'/Random_'+str(dim)+'x'+str(dim)+'dim_'+str(num)+'part_'+str(temp+500)+'.csv'
            shapefile = 'GridShapefiles/'+str(dim)+'x'+str(dim)+'/'+str(dim)+'x'+str(dim)+'.shp'

            # Desired file name for betapoint plan and betapoint figure
            filename_mid = 'Betapoint_'+str(dim)+'x'+str(dim)+'_'+str(num)+'_Beta'+str(beta)+'_'+ex+'_'+str(temp)+'and'+str(temp+500)+'.csv'
            figname_mid = 'Betapoint_fig_'+str(dim)+'x'+str(dim)+'_'+str(num)+'_Beta'+str(beta)+'_'+ex+'_'+str(temp)+'and'+str(temp+500)


            # Record start time
            start_cpu = time.process_time()
            start_wall = time.time()

            # Find betapoint
            TD,opt_val,midpoint_plan_assignment,new_A,actual_beta = midpoint.TD_midpoint(shapefile,filename_A,filename_B,balance_epsilon,
                                                                                         warm_start_LS,warm_start_Init,compact_bool,
                                                                                         col_weight,beta,grid_bool)

    
            cpu_time = round(time.process_time()-start_cpu,5)
            wall_time = round(time.time()-start_wall,5)

            # Report run time
            print('Runtime (cpu): ',cpu_time,' seconds')
            print('Runtime (wall): ',wall_time,' seconds')

            # Write results to file
            writer.writerow([filename_A,filename_B,str(TD),cpu_time,wall_time])

            # Output and visualize the result
            midpoint.output_and_visualize_result(midpoint_plan_assignment,new_A,filename_B,shapefile,filename_mid,figname_mid)

    
    
        outfile.close()
```
